```python
# importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# lendo o arquivo 
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
df = df.drop(columns = 'id', axis=1)
df = df.drop(columns = 'Unnamed: 32', axis=1)
```

```python
df.head()
```

```python
#verificando valores únicos
df['diagnosis'].unique()
```

```python
df['diagnosis'] = df['diagnosis'].map({'M' : 1, 'B' : 0})
df.head()
```

```python
# creating classes. last number not included 
mean_features = list(df.columns[1:11])
se_features = list(df.columns[11:21])
worst_features = list(df.columns[21:31])       
```

```python
# adding the 'diagnosis' column to each to see the correlation
mean_features.append('diagnosis')
se_features.append('diagnosis')
worst_features.append('diagnosis')
```

```python
# now, using 'corr' to look for the correlation
correlation = df[worst_features].corr()
correlation
```

```python
# starting to train the model 
```

```python
# getting the features with the highest correlation
prediction_vars = ['radius_mean', 'perimeter_mean', 'area_mean', 'compactness_mean',
                   'concavity_mean', 'concave points_mean', 'radius_se', 'area_se',
                  'radius_worst', 'perimeter_worst', 'area_worst' ]
```

```python
# now, we are going to train and test 
from sklearn.model_selection import train_test_split

# random to get the seed
train, test = train_test_split(df, test_size=0.15, random_state=1)
```

```python
train_x = train[prediction_vars]
train_y = train['diagnosis'] # the column we want to predict

test_x = test[prediction_vars]
test_y = test['diagnosis']
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
model = RandomForestClassifier()
```

```python
model.fit(train_x, train_y)
```

```python
predictions = model.predict(test_x)
```

```python
# using confusion matrix to show false positivies and false negativies
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y, predictions)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
# checking the precision score of the model
precision = precision_score(test_y, predictions)
print(f'The precision is {precision:.2}')

# recall is the real positivies rate of the model
recall = recall_score(test_y, predictions)
print(f'The recall is {recall:.2}')
```

```python
# reporting the accuracy 
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(test_y, predictions)
print(f'The accuracy is {accuracy:.2}')
```
